## Regresiones lineales simples

Seaborn provee una interfaz conveniente para generar gráficas complejas orientadas a usos estadísticos. Una de las cosas más simples que se puede realizar en Seaborn es ajustar y visualizar regresiones lineales simples entre dos variables usando sns.lmplot()

Seaborn requiere que se pasen los DataFrames directamente al gráfico y que se refieran las columnas por nombre. 


```python
# Importar módulos para graficar
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
auto = pd.read_csv('./dataset/auto-mpg.csv')
# Graficar regresiones lineales simples entre variables.
sns.lmplot(x='weight', y='hp', data=auto)

plt.show()
```

```python
sns.residplot(x='hp', y='mpg', data=auto, color='green')

plt.show()
```

```python
tips =sns.load_dataset('tips')
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex',palette='Set1')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='sex')
plt.show()
```

```python
sns.stripplot(y= 'tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day', y='tip', data=tips, hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x= 'total_bill', y= 'tip', data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill', y= 'tip', data=tips,
kind='kde')
plt.show()
```

```python
sns.pairplot(tips)
plt.show()
```

```python
sns.pairplot(tips, hue='sex')
In [8]: plt.show()
```
