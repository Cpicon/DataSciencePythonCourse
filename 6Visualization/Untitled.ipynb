```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
auto =pd.read_csv('./dataset/auto-mpg.csv')



keys=["mpg","cyl","displ","hp","weight","accel","yr","origin","name","color","size","marker"]

auto.head()
sns.lmplot(x='weight',y='hp',data=auto)

plt.show()
```

```python
tips=sns.load_dataset('tips')
sns.lmplot(x='total_bill',y='tip',data=tips, hue='sex',palette='Set1')
plt.show()
```

```python
print(tips)
```

```python
sns.residplot(x='weight',y='hp',data=auto,color='green')
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='time')
plt.show()
```

```python
sns.stripplot(y= 'tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day',y= 'tip', data=tips)
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.stripplot(x='day',y= 'tip', data=tips,hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
sns.swarmplot(x='day',y= 'tip', data=tips,hue='sex')
plt.ylabel('tip ($)')
plt.show()
```

```python
plt.subplot(1,2,1)
sns.boxplot(x='day',y='tip',data=tips)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip',data=tips)
plt.ylabel('tip ($)')
plt.tight_layout()
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```
